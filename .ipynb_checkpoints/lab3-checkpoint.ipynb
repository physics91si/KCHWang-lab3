```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
def parse_sample(sample, words_only = True, sort_only = True):
    """Parse a string into a list of words."""
    sample = sample.replace("-", " ")
    if words_only:
        sample = "".join(c for c in sample 
                         if c not in ('.', ',', ':', ';', '"', '\'', '?', '!', '&', '[', ']')).lower()
    mylist = sample.split()
    if sort_only:
        mylist.sort()   
    return mylist
```

```python
# Test parse_sample
sample = load_sample()
words_only = True
sort_only = True
parse_sample(sample, words_only, sort_only)
```

```python
def count_freq(lst):
    """Return a dict of word frequencies given a list of words.
    
    Arguments:
    -- lst: a list of strings
    
    Returns:
    -- freq_dict: a dictionary of word: frequency pairs. The keys are strings
       containing the words, and the values are integers indicating how many
       times each word appears.
    
    Example:
    >>> count_freq(['time', 'after', 'time'])
    {'time': 2, 'after': 1}
    """
    from collections import Counter
    return dict(Counter(lst))

def mean(lst):
    """Return the arithmetic mean of a list of numbers."""
    import numpy as np
    return np.mean(lst)

def stdev(lst):
    """Return the standard deviation of a list of numbers."""
    import numpy as np
    return np.std(lst)

def median(lst):
    """Return the median of a list of numbers."""
    import numpy as np
    return np.median(lst)
```

```python
count_freq(['time', 'after', 'time'])
```

```python
mean([15, 18, 2, 36])
```

```python
stdev([1, 2, 3])
```

```python
median([4, 2, 5, 1, 3])
```

```python
def print_stats(lst):
    """Print common statistics about the word frequencies of a list of words."""
    freq = list(count_freq(lst).values())
    print "mean of word frequencies:", mean(freq)
    print "standard deviation of word frequencies:", stdev(freq)
    print "median of word frequencies:", median(freq)
```
